```python
import pandas as pd
from pathlib import Path
import pickle
import yfinance as yf
```

```python
app_data_location = Path(".").resolve().parent.joinpath('can_app').joinpath('canadian_data.sav')
app_data = {}
```

Read excel file as dataframe

```python
df = pd.read_excel(Path(".").resolve().parent.joinpath('data').joinpath(r'CI Canadian All Cap Equity Income - Dec 31 2021 Holdings Report.xlsx'))
df
```

Create yahoo finance tickers

```python
df['stripped_ticker'] = df.apply(lambda x: x['Symbol'] if str(x['Minor Asset Code']) not in ('Common Stock', 'Real Estate Investment Trust') else x['Symbol'].split(" ")[0], axis='columns')
df['yahoo_ticker'] = df.apply(lambda x: x['stripped_ticker'] + '.TO' if x['Country Code'] == 'CAN' else x['stripped_ticker'] + '.L' if x['Country Code'] == 'GBR' else x['stripped_ticker'], axis='columns')
df
```

Address special characters

```python
df.loc[df['Symbol'] == 'AP-U CN', 'yahoo_ticker'] = 'AP-UN.TO'
df.loc[df['Symbol'] == 'BAM/A CN', 'yahoo_ticker'] = 'BAM-A.TO'
df
```

Fetch Yahoo finance data

```python

tickers = yf.Tickers(" ".join(df[df['Minor Asset Code'].isin(['Common Stock', 'Real Estate Investment Trust'])]['yahoo_ticker'].to_list()))
```

```python
market_cap_dict = {}
for key, value in tickers.tickers.items():
    market_cap_dict[key] = [tickers.tickers[key].info['marketCap']]

market_cap_dict
```

```python
market_cap_df = pd.DataFrame(market_cap_dict)
market_cap_df
```

```python
market_cap_df = market_cap_df.stack().reset_index().rename(columns={0: 'market_cap', 'level_1': 'ticker'}).drop(columns=['level_0'])
market_cap_df
```

```python
df = pd.merge(df, market_cap_df, how='left', left_on='yahoo_ticker', right_on='ticker')
df
```

```python
df['weight'] = df['Market Value'] / df['Market Value'].sum()
df
```

```python
top_holding_df = df.sort_values(by='weight', ascending=False).head(10)
top_holding_df
```

```python
app_data['top_holding_df'] = top_holding_df
```

```python
ydata = yf.download("^GSPTSE ^SP1500", start="2022-01-01", end="2022-05-31")
ydata['Adj Close']
```

```python
country_index_df = ydata['Adj Close'].pct_change().fillna(0)
country_index_df
```

```python
country_index_df['usa-cad'] = country_index_df['^SP1500'] - country_index_df['^GSPTSE']
country_diff_df = (country_index_df['usa-cad'] + 1).cumprod() - 1
country_diff_df = country_diff_df.reset_index()
country_diff_df
```

```python
app_data['country_diff_df'] = country_diff_df
```

```python
country_exp_df = df.groupby(by='Country Code')['weight'].sum().reset_index()
country_exp_df
```

```python
app_data['country_exp_df'] = country_exp_df
```

```python
ydata = yf.download("VIGAX VTV", start="2022-01-01", end="2022-05-31")
ydata['Adj Close']
```

```python
growth_index_df = ydata['Adj Close'].pct_change().fillna(0)
growth_index_df
```

```python
growth_index_df['grow-val'] = growth_index_df['VIGAX'] - growth_index_df['VTV']
grow_diff_df = (growth_index_df['grow-val'] + 1).cumprod() - 1
grow_diff_df = grow_diff_df.reset_index()
grow_diff_df
```

```python
app_data['grow_diff_df'] = grow_diff_df
```

```python
ydata = yf.download("VDE VGT", start="2022-01-01", end="2022-05-31")
ydata['Adj Close']
```

```python
sector_df = ydata['Adj Close'].pct_change().fillna(0)
sector_df
```

```python
sector_df['tech-ene'] = sector_df['VGT'] - sector_df['VDE']
sector_diff_df = (sector_df['tech-ene'] + 1).cumprod() - 1
sector_diff_df = sector_diff_df.reset_index()
sector_diff_df
```

```python
app_data['sector_diff_df'] = sector_diff_df
```

```python
pickle.dump(app_data, open(app_data_location, 'wb'))
```
