```python
import pandas as pd
from pathlib import Path
import pickle
import yfinance as yf
```

```python
app_data_location = Path(".").resolve().parent.joinpath('global_app').joinpath('global_data.sav')
app_data = {}
```

```python
df = pd.read_excel(Path(".").resolve().parent.joinpath('data').joinpath(r'CI Select Global Equity - Dec 31, 2021 Holdings Report.xlsx'))
df
```

```python
df['weight'] = df['Market Value'] / df['Market Value'].sum()
df
```

```python
top_holding_df = df.sort_values(by='weight', ascending=False).head(10)
top_holding_df
```

```python
app_data['top_holding_df'] = top_holding_df
```

```python
ydata = yf.download("VCR VGT", start="2022-01-01", end="2022-05-31")
ydata['Adj Close']
```

```python
sector_df = ydata['Adj Close'].pct_change().fillna(0)
sector_df
```

```python
sector_df['tech-cons'] = sector_df['VGT'] - sector_df['VCR']
sector_diff_df = (sector_df['tech-cons'] + 1).cumprod() - 1
sector_diff_df = sector_diff_df.reset_index()
sector_diff_df
```

```python
app_data['sector_diff_df'] = sector_diff_df
```

```python
pickle.dump(app_data, open(app_data_location, 'wb'))
```
